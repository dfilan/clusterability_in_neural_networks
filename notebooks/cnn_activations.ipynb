```python
clustering_results = {}

model_tags = ('CNN-MNIST', 'CNN-MNIST+DROPOUT', 'CNN-MNIST+L1REG', 'CNN-MNIST+MOD-INIT',
              'CNN-FASHION', 'CNN-FASHION+DROPOUT', 'CNN-FASHION+L1REG', 'CNN-FASHION+MOD-INIT',
              'CNN-CIFAR10', 'CNN-CIFAR10+DROPOUT', 'CNN-CIFAR10+L1REG',
              #'CNN-CIFAR10+MOD-INIT'
              )
cnn_model_paths = {model_tag: get_model_path(model_tag) for model_tag in model_tags}

n_clust = 12
n_samples = 100
```
