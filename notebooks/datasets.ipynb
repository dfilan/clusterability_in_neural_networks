```python
import pickle
import numpy as np

data = pickle.load(open('../datasets/mnist-cifar10-separated.pckl', 'rb'))
X = data['X_train']
y = data['y_train']

indices = np.arange(len(y))

samples = np.random.choice(indices, 16)
Xs = X[samples]
ys = y[samples]
```

```python
# https://stackoverflow.com/questions/46615554/how-to-display-multiple-images-in-one-figure-correctly/46616645

import numpy as np
import matplotlib.pyplot as plt

w = h = 28
fig = plt.figure(figsize=(9, 13))
columns = 4
rows = 4

# ax enables access to manipulate each of subplots
axes = []

for i in range(columns*rows):
    img = Xs[i]
    ax = fig.add_subplot(rows, columns, i+1) 
    ax.set_title(ys[i], fontsize=30)
    ax.axis('off')
    plt.imshow(img, cmap='gray', vmin=0, vmax=255)

plt.tight_layout()
plt.savefig('../../../Downloads/mix-mnist-cifar10-separated.png', dpi=300)
```
