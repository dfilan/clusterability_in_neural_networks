# ImageNet Network Clustering Analysis

```python
# https://github.com/qubvel/classification_models

import sys
sys.path.append('..')
from src.spectral_cluster_model import run_clustering_imagenet
from src.visualization import draw_clustered_net_imagenet
```

```python
simple_nets = ['resnet18', 'resnet34', 'vgg16', 'vgg19']
complex_nets = ['resnet50', 'inceptionv3', 'mobilenetv2']
```

```python
n_clust = 16

for net in simple_nets:

    results = run_clustering_imagenet(net, num_clusters=n_clust, num_samples=64,
                                      eigen_solver='amg', num_workers=4)
    print(net)
    print('ncut:', results['ncut'])
    print('n_samples:', results['n_samples'])
    print('mean:', results['mean'])
    print('stdev:', results['stdev'])
    print('z_score:', results['z_score'])
    print('percentile:', results['percentile'])
    print()

    draw_clustered_net_imagenet(results, n_clusters=n_clust)
```

```python
for net in complex_nets:

    results = run_clustering_imagenet(net, num_clusters=n_clust, num_samples=8,
                                      eigen_solver='arpack', num_workers=4)
    print(net)
    print('ncut:', results['ncut'])
    print('n_samples:', results['n_samples'])
    print('mean:', results['mean'])
    print('stdev:', results['stdev'])
    print('z_score:', results['z_score'])
    print('percentile:', results['percentile'])
    print()

    draw_clustered_net_imagenet(results, n_clusters=n_clust)
```
