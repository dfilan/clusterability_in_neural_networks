```python
import sys
sys.path.append('..')
import os
import pickle
os.environ['CUDA_VISIBLE_DEVICES'] = ''
import matplotlib.pylab as plt
import seaborn as sns
from src.spectral_cluster_model import get_ncuts_random_init
```

```python
nets = ['MLP', 'Small CNN', 'VGG CNN']
clust_init_nets = ['MLP Clust. Init', 'Small CNN Clust. Init', 'VGG CNN Clust. Init']
colors = ['firebrick', 'darkgreen', 'mediumblue']
clust_init_colors = ['lightsalmon', 'lightgreen', 'cornflowerblue']
n_trials = 100
results = {}
data_savename = '../results/random_init_data.pkl'
fig_savename = '../results/random_inits.png'
```

```python
for net in nets:
    results[net] = {}
    for mod_init in [False, True]:
        results[net][mod_init] = get_ncuts_random_init(net, n_trials=n_trials, mod_init=mod_init)
with open(data_savename, 'wb') as f:
    pickle.dump(results, f)
```

```python
with open(data_savename, 'rb') as f:
    results = pickle.load(f)
    
fix, ax = plt.subplots(nrows=1, ncols=1, sharey=True, figsize=(12, 8))

for i, net in enumerate(results.keys()):
    sns.distplot(results[net][False], ax=ax, color=colors[i], hist_kws={'alpha': 0.8})
    sns.distplot(results[net][True], ax=ax, color=clust_init_colors[i], hist_kws={'alpha': 0.8})
    
plt.title('N-cuts of n=100 Randomly Initialized Networks, k=12', fontsize=13)
plt.legend(['MLP', 'MLP Clust. Init', 
            'Small CNN', 'Small CNN Clust. Init', 
            'VGG CNN', 'VGG CNN Clust. Init'], fontsize=13)
    
plt.savefig(fig_savename)
```

```python
savename_beginning = '../results/random_inits_'
net_saves = ['mlp', 'small_cnn', 'vgg_cnn']
clust_init_net_saves = ['mlp_clust_init', 'small_cnn_clust_init', 'vgg_cnn_clust_init']

with open(data_savename, 'rb') as f:
    results = pickle.load(f)
    
for i, net in enumerate(results.keys()):
    
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 4))
    sns.distplot(results[net][False], ax=ax, color=colors[i], hist_kws={'alpha': 0.8})
    plt.title(f'{nets[i]} n=100 Random Net N-cuts, k=12', fontsize=12)
    plt.savefig(savename_beginning + net_saves[i] + '.png')
    plt.show()
    
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 4))
    sns.distplot(results[net][True], ax=ax, color=clust_init_colors[i], hist_kws={'alpha': 0.8})
    plt.title(f'{clust_init_nets[i]} n=100 Random Net N-cuts, k=12', fontsize=12)
    plt.savefig(savename_beginning + clust_init_net_saves[i] + '.png')
    plt.show() 
```
